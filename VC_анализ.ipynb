```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
vc = pd.read_feather('all_vc.feather')
```

```python
ind = ''
with open('output.txt', 'r') as f:
    for line in f:
        ind += f.readline()

ind = ind.replace('[', ', ')
ind = ind.replace('][', ', ')
ind = ind.replace(']\n','')
ind = ind.replace(']', ', ')

lst = [0] * len(vc)
for i in set(ind.split(', ')):
    try:
        lst[int(i)] = 1
    except:
        print(i)
        
vc['fintech_text'] = lst
```

```python
vc.dropna(inplace=True)
```

```python
def lencom(x):
    return len(x.split('~#')) - 1
```

```python
def getdate(date):
    a = date.split('-')
    if a[0] == 'в':
        return 2020 * 100 + int(a[1])
    elif len(a[0]) > 4:
        return 202000 + 3
    return int(a[0]) * 100 + int(a[1])
```

```python
vc['date_for_sort'] = vc['date'].apply(getdate)
```

```python
def realdate(date):
    return str(date // 100) + '-' + str(date % 100)
```

```python
vc['date'] = vc['date_for_sort'].apply(realdate)
```

```python
vc.sort_values(by=['date_for_sort'], inplace=True)
```
